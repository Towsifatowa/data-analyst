 Automated File Sorter in File Explorer

```python
import os, shutil
```

```python
path = r"C:/Users/LENOVO/Documents/DATA analytics/python(alex)/"
```

```python
file_name = os.listdir(path)
```

```python
import os

folder_name = ['cvc files', 'image files', 'text files']

for name in folder_name:
    full_path = os.path.join(path, name)
    if not os.path.exists(full_path):
        print(full_path)
        os.makedirs(full_path)
```

```python
for file in file_name:
    if ".csv" in file and not os.path.exists(path + "csv files/" + file):
        shutil.move(path + file, path + "csv files/" + file)

    elif ".jpeg" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)

    elif ".txt" in file and not os.path.exists(path + "text files/" + file):
        shutil.move(path + file, path + "text files/" + file)

    else:
        print("There are files in this path that were not moved")
```
